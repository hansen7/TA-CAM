```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, matplotlib.pylab as pylab, seaborn as sns

params = {'legend.fontsize': 'x-large',
          'figure.figsize': (16, 3),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
sns.set()
pylab.rcParams.update(params)
```

### Q6(c) of Example Paper 2

```python
k = np.arange(-4, 4.1, 1)
f = np.arange(-2.1, 2.1, .01)
a, sigma, sigmax = 0.9, 1, 5.26

Rx = a**np.abs(k) * sigmax
Sx = sigma/(1 + a**2 - 2*a*np.cos(2*np.pi*f))

fig = plt.figure()

plt.subplot(121)
plt.xlabel('$k$')
plt.ylabel('$R_X(k)$')
plt.stem(k, Rx, use_line_collection=True)

plt.subplot(122)
plt.xlabel('$f$')
plt.ylabel('$S_X(f)$')
plt.plot(f, Sx)

plt.tight_layout()
plt.savefig(r'./fig/ep2q6c.png', dpi=200)
```

### Q8(c) of Example Paper 2

```python
x = np.arange(-4, 4.1, 1)
q = 2

Rx = (np.heaviside(q-x, 1) * (q+1-x)) * np.heaviside(x, 1)

fig = plt.figure()

plt.subplot(111)
plt.xlabel('$x(q=2)$')
plt.ylabel('$R_X(x)$')
plt.stem(k, Rx, use_line_collection=True)

# plt.subplot(122)
# plt.xlabel('$f$')
# plt.ylabel('$S_X(f)$')
# plt.plot(f, Sx)

plt.tight_layout()
plt.savefig(r'./fig/ep2q8c.png', dpi=200)
```

### Q8(d) of Example Paper 2

```python
f = np.arange(-4, 4.1, .01)
sigma, b1, b2 = 1, 1, 2

Hf = sigma * (1 + b1**2 + b2**2 + 2*b1*(1+b2)*np.cos(2*np.pi*f) + 2*b2*np.cos(4*np.pi*f))

fig = plt.figure()

plt.subplot(111)
plt.xlabel('$f\,\,(q=2, \sigma^2=1, b_1=1, b_2=2)$')
plt.ylabel('$H_F(f)$')
plt.plot(f, Hf)

plt.tight_layout()
plt.savefig(r'./fig/ep2q8d.png', dpi=200)
```
